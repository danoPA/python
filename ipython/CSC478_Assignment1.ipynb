```python
# End of question 3

# Start of question 4:
# Discretize the age attribute into 3 categories (corresponding to "young", "mid-age", and "old").
# [Do not change the original age attribute in the table.]
age_bins = pd.qcut(vstable.age, [0, .33, .66, 1], labels=["young", "mid-age", "old"], retbins=True)
age_bins
```

```python
# End of question 4

# Start of question 5: Use Min-Max Normalization to transform the values of all numeric attributes 
# in the original table (income, age, children) onto the range 0.0-1.0.
vs_numeric = vstable[["income","age","children"]]
vs_norm = (vs_numeric - vs_numeric.min()) / (vs_numeric.max() - vs_numeric.min())
vs_norm.head(10)
```

```python
# End of question 5

# Start of question 6: Convert the table (after normalization in part 5) into the standard spreadsheet format. 
# Note that this requires converting each categorical attribute into multiple binary ("dummy") attributes 
# (one for each values of the categorical attribute) and assigning binary values corresponding to the presence 
# or not presence of the attribute value in the original record). Then numeric attributes should remain unchanged. 
# Save this new table into a file called bank_numeric.csv and submitit along with your assignment. 
# [Hint: you might consider using the get_dummies for Pandas data frames.]
vs_ssf = pd.get_dummies(vstable)
vs_norm_all = (vs_ssf - vs_ssf.min()) / (vs_ssf.max() - vs_ssf.min())
vs_norm_all.head(10).T
```

```python
vs_norm_all.to_csv("bank_numeric.csv", float_format="%1.2f")
```
